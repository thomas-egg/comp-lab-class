```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import os
```

# Plots of 1AKI MD Run

```python
# Load energy minimization data
file_path = '/home/tje3676/comp-lab-class/comp-lab-class/Week3/Data/potential.xvg'
energy_min = np.genfromtxt(file_path, skip_header=24)
```

```python
# Plot energy vs. steps
steps = energy_min[:,0]
energy = energy_min[:,1]

plt.plot(steps, energy)
plt.title('Energy Minimization')
plt.xlabel('Number of steps')
plt.ylabel('Potential Energy (kJ/mol)')
plt.show()
```

```python
# Load temperature equilibration data
file_path = '/home/tje3676/comp-lab-class/comp-lab-class/Week3/Data/temperature.xvg'
equil = np.genfromtxt(file_path, skip_header=24)

# Plot equilibration
time = equil[:,0]
temp = equil[:,1]

plt.plot(time, temp)
plt.title('Temperature vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.show()
```

```python
# Load pressure equilibration data
file_path = '/home/tje3676/comp-lab-class/comp-lab-class/Week3/Data/pressure.xvg'
equil = np.genfromtxt(file_path, skip_header=24)

# Plot equilibration
time = equil[:,0]
pressure = equil[:,1]

# Pressure rolling average 
rolling_avg = []
for i in range(10, len(pressure)):
    rolling_avg.append(np.mean(pressure[i-10:i]))

plt.plot(time, pressure)
plt.plot(time[10:], rolling_avg)
plt.title('Pressure vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (K)')
plt.legend(['Pressure', 'Pressure (10 ps avg)'])
plt.show()
```
