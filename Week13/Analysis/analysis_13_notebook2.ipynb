```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
rc_data = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog')
```

```python
# Plot data
plt.title('RC vs Time')
plt.xlabel('Time (fs)')
plt.ylabel('$\sqrt{d_{5663}^2+d_{5675}^2} - \sqrt{d_{5670}^2+d_{5671}^2}$')
plt.plot(rc_data[189:,0]/10, rc_data[189:,1])
plt.show()
```

```python
# Import new data
meta_rc_data = np.loadtxt('../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog')
```

```python
# Plot data
plt.title('RC vs Time (Metadynamics)')
plt.xlabel('Time (fs)')
plt.ylabel('$\sqrt{d_{5663}^2+d_{5675}^2} - \sqrt{d_{5670}^2+d_{5671}^2}$')
plt.plot(meta_rc_data[:,0]/10, meta_rc_data[:,1])
plt.show()
```

```python
# Load freqhills data
freq_hills_data = np.loadtxt('../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog')
```

```python
# Plot data
plt.title('RC vs Time (Metadynamics)')
plt.xlabel('Time (fs)')
plt.ylabel('$\sqrt{d_{5663}^2+d_{5675}^2} - \sqrt{d_{5670}^2+d_{5671}^2}$')
plt.plot(rc_data[189:,0]/10, rc_data[189:,1], label='monitor')
plt.plot(freq_hills_data[:,0]/10, freq_hills_data[:,1], label='height=1.5, width=0.5, freq=20')
plt.plot(meta_rc_data[:,0]/10, meta_rc_data[:,1], label='height=1.5, width=0.5, freq=100')
plt.legend()
plt.show()
```
