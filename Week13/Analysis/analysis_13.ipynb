# Week 13 Analysis

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
nvt_data = np.loadtxt('../Outputs/NVT-1.ener')
npt_data = np.loadtxt('../Outputs/NPT-1.ener')
```

```python
# Plot NVT data
plt.title('Potential Energy vs. Time (NVT)')
plt.xlabel('Time (fs)')
plt.ylabel('Energy (a.u.)')
plt.plot(nvt_data[:,1], nvt_data[:,4])
plt.show()
```

```python
# Plot NPT data
plt.title('Potential Energy vs. Time (NPT)')
plt.xlabel('Time (fs)')
plt.ylabel('Energy (a.u.)')
plt.plot(npt_data[:,1], npt_data[:,4])
plt.show()
```
