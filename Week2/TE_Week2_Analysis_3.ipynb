```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import os
```

# Trajectory Visualization

```python
# Load trajectory
traj = md.load_xtc('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='Data/1hz3_T310.start.nowater.gro')
```

End-to-end distance is calculated by taking the distance between residues on either end of protein.

```python
# Save first and last
first = traj.xyz[:,0,:]
last = traj.xyz[:,-1,:]

# Compute distance (distance formula)
distance = np.sqrt(np.sum(np.power((last-first), 2), axis=1))

# Plot distance
plt.plot(traj.time, distance)
plt.title('End-to-end Distance vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.show()
```

```python
# Calculate gyration
gyration = md.compute_rg(traj)

# Plot results
plt.plot(traj.time, gyration, )
plt.plot(traj.time, distance)
plt.title('Distances vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.legend(['$R_{g}$', 'End-to-end'])
plt.show()
```

```python
# End-to-end histogram
hist, bins = np.histogram(distance)
plt.hist(hist, density=True)
plt.show()
```

```python
# Gyration histogram
hist, bins = np.histogram(gyration)
plt.hist(hist, density=True)
plt.show()
```
