# Trp-cage System

This file contains the code to generate figures pertaining to the Week 4 simulation of trp-cage.

```python
# Import libraries
import numpy as np
import pandas as pd
import mdtraj as md
import matplotlib.pyplot as plt
```

```python
# Import trajectory
file_path = '/home/tje3676/comp-lab-class/comp-lab-class/Week4/trp-cage_AMBER/gromacs/trp-cage_center.xtc'
top_file = '/home/tje3676/comp-lab-class/comp-lab-class/Week4/trp-cage_AMBER/2jof.pdb'
traj = md.load_xtc(file_path, top_file)
```

```python
# compute the distance between the first and last atom
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

# compute the distance between the first and last atom
dist = end_end(traj)

#plot distance vs time
plt.title("end-to-end distance vs. Time")
plt.plot(traj.time, dist)
plt.xlabel('Time (ps)')
plt.ylabel('End-to-end Distance (nm)')
```

```python
# Calculate gyration
gyration = md.compute_rg(traj)

# Plot results
plt.plot(traj.time, gyration)
plt.title('Radius of Gyration vs. Time')
plt.xlabel('Time (ps)')
plt.ylabel('$R_g$ (nm)')
plt.show()
```
